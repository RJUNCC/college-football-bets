```python
import polars as pl 
import altair as alt
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions as EC
```

```python
MAIN_URL = "https://www.pro-football-reference.com/teams/"
```

```python
driver = webdriver.Firefox()
driver.get(url=MAIN_URL)
teams = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row] th a')
year_min = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="year_min"]')
year_max = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="year_max"]')
wins = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="wins"]')
losses = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="losses"]')
ties = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="ties"]')
win_loss_perc = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="win_loss_perc"]')
av = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="av"]')
passer = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="passer"]')
rusher = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="rusher"]')
receiver = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="receiver"]')
coachwinner = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="coachwinner"]')
years_playoffs = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="years_playoffs"]')
wins_playoffs = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="wins_playoffs"]')
losses_playoffs = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="losses_playoffs"]')
win_loss_perc_playoffs = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="win_loss_perc_playoffs"]')
championships = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="championships"]')
championships_super_bowl = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="championships_super_bowl"]')
championships_conference = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="championships_conference"]')
championships_division = driver.find_elements(By.CSS_SELECTOR, 'div[id="div_teams_active"] table[id="teams_active"] tr[data-row]:not(.partial_table) td[data-stat="championships_division"]')

teams_list = [elem.text for elem in teams]
year_min_list = [elem.text for elem in year_min]
year_max_list = [elem.text for elem in year_max]
wins_list = [elem.text for elem in wins]
losses_list = [elem.text for elem in losses]
ties_list = [elem.text for elem in ties]
win_loss_perc_list = [elem.text for elem in win_loss_perc]
av_list = [elem.text for elem in av]
passer_list = [elem.text for elem in passer]
rusher_list = [elem.text for elem in rusher]
receiver_list = [elem.text for elem in receiver]
coachwinner_list = [elem.text for elem in coachwinner]
years_playoffs_list = [elem.text for elem in years_playoffs]
wins_playoffs_list = [elem.text for elem in wins_playoffs]
losses_playoffs_list = [elem.text for elem in losses_playoffs]
win_loss_perc_playoffs_list = [elem.text for elem in win_loss_perc_playoffs]
championships_list = [elem.text for elem in championships]
championships_super_bowl_list = [elem.text for elem in championships_super_bowl]
championships_conference_list = [elem.text for elem in championships_conference]
championships_division_list = [elem.text for elem in championships_division]


header_elems = driver.find_elements(By.CSS_SELECTOR, "thead tr th")
header_elems_list = [elem.text for elem in header_elems if elem.text != ""]
driver.close()
```

```python
cols = header_elems_list[2:22]
values = [
    teams_list,
    year_min_list,
    year_max_list,
    wins_list,
    losses_list,
    ties_list,
    win_loss_perc_list,
    av_list,
    passer_list,
    rusher_list,
    receiver_list,
    coachwinner_list,
    years_playoffs_list,
    wins_playoffs_list,
    losses_playoffs_list,
    win_loss_perc_playoffs_list,
    championships_list,
    championships_super_bowl_list,
    championships_conference_list,
    championships_division_list
]
cols
```

```python
temp_dict = {}
for key, val in zip(cols, values):
    temp_dict[key] = val
```

```python
for i in range(len(values)):
    print(len(values[i]), len(cols[0]))
```

```python
df = pl.DataFrame(temp_dict)
df.head()
```

```python
df = df.with_columns(
    pl.col("From").str.to_date("%Y").dt.year(),
    pl.col("To").str.to_date("%Y").dt.year(),
    pl.col("W").cast(pl.Utf8).cast(pl.Int32),
    pl.col("L").cast(pl.Utf8).cast(pl.Int32),
    pl.col("T").cast(pl.Utf8).cast(pl.Int32),
    pl.col("W-L%").cast(pl.Utf8).cast(pl.Float32),
    pl.col("Yr plyf").cast(pl.Utf8).cast(pl.Int32),
    pl.col("W plyf").cast(pl.Utf8).cast(pl.Int32),
    pl.col("L plyf").cast(pl.Utf8).cast(pl.Int32),
    pl.col("Chmp").cast(pl.Utf8).cast(pl.Int32),
    pl.col("SBwl").cast(pl.Utf8).cast(pl.Int32),
    pl.col("Conf").cast(pl.Utf8).cast(pl.Int32),
    pl.col("Div").cast(pl.Utf8).cast(pl.Int32),
)
```
