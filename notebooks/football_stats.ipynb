```python
import polars as pl 
import altair as alt
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions as EC
```

```python
MAIN_URL = "https://www.pro-football-reference.com/teams/"
```

```python
driver = webdriver.Firefox()
driver.get(url=MAIN_URL)
teams = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row] th a')
year_min = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="year_min"]')
year_max = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="year_max"]')
wins = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="wins"]')
losses = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="losses"]')
ties = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="ties"]')
win_loss_perc = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="win_loss_perc"]')
av = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="av"]')
passer = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="passer"]')
rusher = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="rusher"]')
receiver = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="receiver"]')
coachwinner = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="coachwinner"]')
years_playoffs = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="years_playoffs"]')
wins_playoffs = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="wins_playoffs"]')
losses_playoffs = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="losses_playoffs"]')
win_loss_perc_playoffs = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="win_loss_perc_playoffs"]')
championships = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="championships"]')
championships_super_bowl = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="championships_super_bowl"]')
championships_conference = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="championships_conference"]')
championships_division = driver.find_elements(By.CSS_SELECTOR, 'tr[data-row]:not(.partial_table) td[data-stat="championships_divison"]')

teams_list = [elem.text.split() for elem in teams]
year_min_list = [elem.text.split() for elem in year_min]
year_max_list = [elem.text.split() for elem in year_max]
wins_list = [elem.text.split() for elem in wins]
losses_list = [elem.text.split() for elem in losses]
ties_list = [elem.text.split() for elem in ties]
win_loss_perc_list = [elem.text.split() for elem in win_loss_perc]
av_list = [elem.text.split() for elem in av]
passer_list = [elem.text.split() for elem in passer]
rusher_list = [elem.text.split() for elem in rusher]
receiver_list = [elem.text.split() for elem in receiver]
coachwinner_list = [elem.text.split() for elem in coachwinner]
years_playoffs_list = [elem.text.split() for elem in years_playoffs]
wins_playoffs_list = [elem.text.split() for elem in wins_playoffs]
losses_playoffs_list = [elem.text.split() for elem in losses_playoffs]
win_loss_perc_playoffs_list = [elem.text.split() for elem in win_loss_perc_playoffs]
championships_list = [elem.text.split() for elem in championships]
championships_super_bowl_list = [elem.text.split() for elem in championships_super_bowl]
championships_conference_list = [elem.text.split() for elem in championships_conference]
championships_division_list = [elem.text.split() for elem in championships_division]


header_elems = driver.find_elements(By.CSS_SELECTOR, "thead tr th")
driver.close()
```
